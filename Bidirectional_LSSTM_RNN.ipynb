```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Bidirectional

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
## vocabulary size
voc_size=9000
```

```python
messages=X.copy()
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
import re
from nltk.corpus import stopwords

ps = PorterStemmer()
corpus = []

# Check the available columns in messages
print("Columns in messages:", messages.columns)

for index, row in messages.iterrows():
    # Print the index and title for debugging
    print("Processing index:", index)
    print("Title:", row['title'])

    review = re.sub('[^a-zA-Z]', ' ', row['title'])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Find the sentence with the maximum length in the corpus
max_length_sentence = max(corpus, key=len)
max_length_index = corpus.index(max_length_sentence)

print("Maximum length sentence:", max_length_sentence)
print("Length:", len(max_length_sentence.split()))
print("Index in the corpus:", max_length_index)
```

```python
# onehhot encoding
## vocabulary size
voc_size=9000
```

```python
onehot_repr=[one_hot(words , voc_size) for words in corpus]
```

```python
onehot_repr
```

```python
# paddding
```

```python
sent_length=50
```

```python
embedded_docs= pad_sequences(onehot_repr , padding='pre' , maxlen=sent_length)
embedded_docs
```

```python
# Embedding
# Creaet teh models
```

```python
# each and every vector is get represent by 40 different vector feature represnttaion (dimention)
# Select the dimension size

embedding_vector_features=40
```

```python
# Create biderectional LSTM model
```

```python
from tensorflow.keras.layers import Dropout
embedding_vector_features=40  # dim

model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(Bidirectional(LSTM(100)))               # <----- Bidirectional
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0) ##AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

def preprocess_text(text):
    ps = PorterStemmer()

    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    text = [ps.stem(word) for word in text.split() if word not in stopwords.words('english')]
    text = ' '.join(text)
    return text

new_headline = "NASA Admits to Hiring Humans to Dress as Aliens for Mars Missions"

preprocessed_headline = preprocess_text(new_headline)


```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Convert to one-hot representation
onehot_repr_new = [one_hot(preprocessed_headline, voc_size)]

# Pad sequence
padded_sequence_new = pad_sequences(onehot_repr_new, padding='post', maxlen=sent_length)

# Make prediction
prediction = model.predict(np.array(padded_sequence_new))

threshold = 0.5
if prediction > threshold:
    print("The news is likely to be fake!")
else:
    print("The news is likely to be genuine.")
```
